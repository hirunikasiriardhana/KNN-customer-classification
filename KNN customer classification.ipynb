```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing# for data normalize
%matplotlib inline
```

```python
df = pd.read_csv('teleCust1000t.csv')
df.head()
```

```python
df.columns, df.shape #columns set
```

```python
df['custcat'].value_counts()
```

```python
X = df.drop(['custcat'], axis = 1)#drop custcat column
X.head()
```

```python
y = df['custcat'] #create y variable and put custcat column
y.head()
```

```python
df.hist(column='income', bins=50)
```

# Normalize data

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
```

```python
X[0:1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics #, we can display a graph using the metrics library 
```

## Definee and train the KNN model

```python
k = 4
#Train Model 
model = KNeighborsClassifier(n_neighbors = k)
model.fit(X_train,y_train)
```

### Get predictions

```python
Pred_y = model.predict(X_test)
print("Accuracy of model at K=4 is",metrics.accuracy_score(y_test, Pred_y))
```

```python
Pred_y
```

```python
Ks = 40
error_rate = []
# Will take some time
for i in range(1,40):
 
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(X_train,y_train)
 pred_i = knn.predict(X_test)
 error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')

print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```

```python
acc = []
# Will take some time
from sklearn import metrics
for i in range(1,100):
    neigh = KNeighborsClassifier(n_neighbors = i).fit(X_train,y_train)
    yhat = neigh.predict(X_test)
    acc.append(metrics.accuracy_score(y_test, yhat))

plt.figure(figsize=(10,6))
plt.plot(range(1,100),acc,color = 'blue',linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')

print("Maximum accuracy:-",max(acc),"at K =",acc.index(max(acc)))
```
